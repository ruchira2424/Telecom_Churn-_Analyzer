```python
cols_to_drop=["State","Area code"]
train_df=train_df.drop(columns=cols_to_drop)
```

```python
test_df=test_df.drop(columns=cols_to_drop)
```

```python
test_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
train_df["International plan"]=le.fit_transform(train_df["International plan"])
train_df["Voice mail plan"]=le.fit_transform(train_df["Voice mail plan"])
test_df["International plan"]=le.fit_transform(test_df["International plan"])
test_df["Voice mail plan"]=le.fit_transform(test_df["Voice mail plan"])
```

```python
train_df.head()
```

```python
train_df["Churn"] = train_df["Churn"].astype(int)
test_df["Churn"] = test_df["Churn"].astype(int)
```

```python
x_train = train_df.drop("Churn", axis=1)  # features
y_train = train_df["Churn"]                # target

x_test = test_df.drop("Churn", axis=1)    # features
y_test = test_df["Churn"]                  # target
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=5000)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test_scaled)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,f1_score,classification_report
```

```python
print("Accuracy",accuracy_score(y_test,y_pred))
print("Precision",precision_score(y_test,y_pred))
print("Recall",recall_score(y_test,y_pred))
print("confusion_matrix\n",confusion_matrix(y_test,y_pred))
print("F1 score",f1_score(y_test,y_pred))
print("\nclassification report\n",classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Get prediction probabilities
y_probs = model.predict_proba(x_test_scaled)[:, 1]

# Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
auc_score = roc_auc_score(y_test, y_probs)

# Plot
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {auc_score:.2f})")
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line for random model
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
